```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set styles for better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
file_path = "/content/sample_-_superstore.xls"
```

```python
df = pd.read_excel(file_path)
```

```python
print("Dataset Overview:")
print(df.info())
print("\nSample Data:")
print(df.head())
```

```python
df = df.dropna() # Drop any missing values for simplicity
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.values, y=category_sales.index, palette="viridis")
plt.title("Total Sales by Product Category")
plt.xlabel("Sales ($)")
plt.ylabel("Category")
plt.show()
```

```python
#  Monthly Sales Trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month-Year'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month-Year')['Sales'].sum()
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales ($)")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
sns.barplot(x=region_sales.values, y=region_sales.index, palette='coolwarm')
plt.title("Sales by Region")
plt.xlabel("Total Sales ($)")
plt.ylabel("Region")
plt.show()
```

```python
#  Step 8: Profit vs Sales by Sub-Category
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Sub-Category')
plt.title("Sales vs Profit by Sub-Category")
plt.show()
```

```python
# Top 10 States by Sales
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
state_sales.plot(kind='barh', color='teal')
plt.title("Top 10 States by Sales")
plt.xlabel("Sales ($)")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Summary
print("Summary Insights:")
print("""
- Technology is the highest revenue-generating category.
- Sales peaked during holiday months (NovDec).
- West and East regions perform better in terms of total sales.
- Profitability is not always aligned with high sales (e.g., some sub-categories are loss-making).
- California and New York are top performing states.
""")
```
